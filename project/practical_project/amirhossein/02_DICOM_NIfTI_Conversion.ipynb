```python
import os
import SimpleITK as sitk
import pandas as pd

# Define paths
linked_metadata_path = "linked_metadata.csv"  # Your linked metadata file
output_nifti_dir = "NIfTI_Converted"  # Output directory for NIfTI files

# Ensure output directory exists
os.makedirs(output_nifti_dir, exist_ok=True)

# Load linked metadata
metadata_df = pd.read_csv(linked_metadata_path)

# Function to convert DICOM to NIfTI
def convert_dicom_to_nifti(dicom_folder, output_nifti_path):
    try:
        reader = sitk.ImageSeriesReader()
        dicom_series = reader.GetGDCMSeriesFileNames(dicom_folder)
        reader.SetFileNames(dicom_series)
        image = reader.Execute()
        sitk.WriteImage(image, output_nifti_path)
        print(f"Converted: {output_nifti_path}")
    except Exception as e:
        print(f"Failed to convert {dicom_folder}: {e}")

# Loop through metadata and convert each patientâ€™s MRI scans
for index, row in metadata_df.iterrows():
    subject_id = row["Subject ID"]
    dicom_path = row["File Location"]  # Path to DICOM folder
    series_desc = row["Series Description"]  # Series name (T1, T2, Segmentation)
    
    # Define output filename
    output_filename = f"{subject_id}_{series_desc.replace(' ', '_')}.nii.gz"
    output_nifti_path = os.path.join(output_nifti_dir, output_filename)
    
    # Convert DICOM to NIfTI
    convert_dicom_to_nifti(dicom_path, output_nifti_path)

print(f"Conversion complete. NIfTI files saved in: {output_nifti_dir}")
```
