```python
import os
import SimpleITK as sitk

# Define input and output directories
nifti_input_dir = "NIfTI_Converted"
nifti_output_dir = "NIfTI_Reoriented"

# Ensure output directory exists
os.makedirs(nifti_output_dir, exist_ok=True)

# Function to reorient NIfTI images using SimpleITK
def reorient_nifti_sitk(input_nifti_path, output_nifti_path):
    try:
        # Load NIfTI image
        image = sitk.ReadImage(input_nifti_path)

        # Convert image to standard RAI orientation
        image = sitk.DICOMOrient(image, "RAI")  # Can be "LPS" or "RAI"

        # Save reoriented image
        sitk.WriteImage(image, output_nifti_path)
        print(f"Reoriented: {output_nifti_path}")

    except Exception as e:
        print(f"Failed to reorient {input_nifti_path}: {e}")

# Loop through all NIfTI files and reorient them
for filename in os.listdir(nifti_input_dir):
    if filename.endswith(".nii.gz"):
        input_nifti_path = os.path.join(nifti_input_dir, filename)
        output_nifti_path = os.path.join(nifti_output_dir, filename)
        reorient_nifti_sitk(input_nifti_path, output_nifti_path)

print(f"Reorientation complete. Files saved in: {nifti_output_dir}")
```
